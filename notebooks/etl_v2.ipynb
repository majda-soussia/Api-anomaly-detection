# **3. PARSING & ENRICHISSEMENT**

### 3.1 Parsing du status code

```python
# Catégorisation des status codes
df["is_success"]   = df["status_code"].between(200, 299).astype(int)
df["is_redirect"]  = df["status_code"].between(300, 399).astype(int)
df["is_error_4xx"] = df["status_code"].between(400, 499).astype(int)
df["is_error_5xx"] = df["status_code"].between(500, 599).astype(int)

# Colonne catégorielle lisible
def categorize_status(code):
    if 200 <= code <= 299: return "success"
    if 300 <= code <= 399: return "redirect"
    if 400 <= code <= 499: return "client_error"
    if 500 <= code <= 599: return "server_error"
    return "unknown"

df["status_category"] = df["status_code"].apply(categorize_status)

print("Status codes parsés")
print(df["status_category"].value_counts())
```

### 3.2 Parsing du user_agent

```python
def classify_agent(ua):
    if pd.isna(ua):
        return "unknown"
    ua = ua.lower()
    bots = ["bot", "crawler", "spider", "curl", "python", "postman", "go-http"]
    if any(b in ua for b in bots):
        return "bot"
    if any(m in ua for m in ["android", "iphone", "mobile", "pixel"]):
        return "mobile"
    if "macintosh" in ua or "windows" in ua:
        return "desktop"
    return "other"

df["client_type"] = df["user_agent"].apply(classify_agent)

print("User agents classifiés")
print(df["client_type"].value_counts())
```

### 3.3 Parsing du path (endpoint)

```python
def classify_path(path):
    if pd.isna(path):
        return "unknown"
    if "/api/" in path:
        return "api"
    if "/static/" in path:
        return "static"
    if any(x in path for x in ["/login", "/register", "/auth"]):
        return "auth"
    if "/admin" in path:
        return "admin"
    return "page"

df["endpoint_type"] = df["path"].apply(classify_path)

# Extraire le nom de l'endpoint API principal
def extract_api_name(path):
    parts = path.strip("/").split("/")
    # /api/v2/payments/process → "payments"
    if len(parts) >= 3 and parts[0] == "api":
        return parts[2] if len(parts) > 2 else parts[-1]
    return "non-api"

df["api_name"] = df["path"].apply(extract_api_name)

print("Endpoints parsés")
```

### 3.4 Extraction des features temporelles

```python
df["hour"] = df["timestamp"].dt.hour
df["minute"] = df["timestamp"].dt.minute
df["dayofweek"]= df["timestamp"].dt.dayofweek      # 0=lundi, 6=dimanche
df["day"] = df["timestamp"].dt.day
df["month"] = df["timestamp"].dt.month
df["is_weekend"] = (df["dayofweek"] >= 5).astype(int)
df["is_business_hours"] = df["hour"].between(8, 18).astype(int)

print("Features temporelles extraites")
df[["timestamp", "hour", "dayofweek", "is_weekend", "is_business_hours"]].head(5)
```

# **4. FEATURE ENGINEERING**

```python
# Indexer par timestamp pour le resample
df_indexed = df.set_index("timestamp").sort_index()

def compute_window_features(group):
    return pd.Series({
        # Volume
        "request_count" : len(group),
        "unique_ips": group["ip_address"].nunique(),

        # Latence
        "avg_response_time"  : group["response_time_ms"].mean(),
        "median_response_time" : group["response_time_ms"].median(),
        "p95_response_time" : group["response_time_ms"].quantile(0.95),
        "p99_response_time": group["response_time_ms"].quantile(0.99),
        "max_response_time" : group["response_time_ms"].max(),
        "std_response_time": group["response_time_ms"].std() if len(group) > 1 else 0,

        # Taux d'erreurs
        "error_rate_5xx" : group["is_error_5xx"].mean(),
        "error_rate_4xx" : group["is_error_4xx"].mean(),
        "success_rate"  : group["is_success"].mean(),

        # Bytes
        "avg_bytes_sent" : group["bytes_sent"].mean(),
        "total_bytes_sent" : group["bytes_sent"].sum(),

        # Types de clients
        "bot_ratio" : (group["client_type"] == "bot").mean(),
        "mobile_ratio"  : (group["client_type"] == "mobile").mean(),
        "desktop_ratio" : (group["client_type"] == "desktop").mean(),

        # Types d'endpoints
        "api_ratio" : (group["endpoint_type"] == "api").mean(),
        "static_ratio": (group["endpoint_type"] == "static").mean(),
        "auth_ratio" : (group["endpoint_type"] == "auth").mean(),

        # Méthodes HTTP
        "get_ratio"  : (group["method"] == "GET").mean(),
        "post_ratio"  : (group["method"] == "POST").mean(),

        # Contexte temporel
        "hour"   : group["hour"].iloc[0],
        "is_weekend" : group["is_weekend"].iloc[0],
        "is_business_hours" : group["is_business_hours"].iloc[0],
    })

features_df = (
    df_indexed
    .groupby(["server_id", pd.Grouper(freq="5min")])
    .apply(compute_window_features)
    .reset_index()
    .dropna(subset=["avg_response_time"])
)
# After you build features_df, add these:
features_df = features_df.sort_values(["server_id", "timestamp"])

# How much did things change compared to last window?
features_df["response_time_change"] = features_df.groupby("server_id")["avg_response_time"].diff()
features_df["request_count_change"] = features_df.groupby("server_id")["request_count"].diff()
features_df["error_rate_change"]    = features_df.groupby("server_id")["error_rate_5xx"].diff()

# Rolling average (last 3 windows = 15 minutes)
features_df["rolling_avg_response"] = (
    features_df.groupby("server_id")["avg_response_time"]
    .transform(lambda x: x.rolling(3, min_periods=1).mean())
)

# How far is this window from recent normal?
features_df["response_vs_rolling"] = (
    features_df["avg_response_time"] - features_df["rolling_avg_response"]
)
print(f"Feature engineering terminé")
print(f"Shape : {features_df.shape}")
features_df.head()
```

# **5. Labelling**

```python
SEUIL_ERROR_5XX   = 0.05    # plus de 5% d'erreurs serveur
SEUIL_P95_LATENCE = 2000    # p95 latence > 2 secondes
SEUIL_TRAFIC_MULT = 3.0     # volume x3 par rapport à la moyenne
mean_requests = features_df["request_count"].mean()
print("SEUILS D'ANOMALIE")
print(f"Taux erreur 5xx : > {SEUIL_ERROR_5XX*100}%")
print(f"Latence p95 : > {SEUIL_P95_LATENCE} ms")
print(f"Moyenne req/5min : {mean_requests:.1f}")
print(f"Seuil pic trafic  : > {mean_requests * SEUIL_TRAFIC_MULT:.1f} req/5min")
```

```python
# Coller ce bloc dans votre notebook après avoir calculé features_df

print("="*50)
print("DIAGNOSTIC DES SEUILS")
print("="*50)

# --- Erreurs 5xx ---
mean_err = features_df["error_rate_5xx"].mean()
std_err  = features_df["error_rate_5xx"].std()
non_zero = features_df[features_df["error_rate_5xx"] > 0]["error_rate_5xx"]

print(f"\n[ERROR 5xx]")
print(f"  mean                = {mean_err:.6f}")
print(f"  std                 = {std_err:.6f}")
print(f"  mean + 2*std        = {mean_err + 2*std_err:.6f}  ← seuil actuel")
print(f"  médiane (non-zero)  = {non_zero.median():.6f}")
print(f"  quantile 95%        = {features_df['error_rate_5xx'].quantile(0.95):.6f}")

# --- Latence P95 ---
mean_lat = features_df["p95_response_time"].mean()
std_lat  = features_df["p95_response_time"].std()

print(f"\n[LATENCE P95]")
print(f"  mean                = {mean_lat:.1f} ms")
print(f"  std                 = {std_lat:.1f} ms")
print(f"  mean + 2*std        = {mean_lat + 2*std_lat:.1f} ms")
print(f"  mean + 3*std        = {mean_lat + 3*std_lat:.1f} ms")
print(f"  quantile 95%        = {features_df['p95_response_time'].quantile(0.95):.1f} ms")

# --- Trafic ---
mean_req = features_df["request_count"].mean()
std_req  = features_df["request_count"].std()

print(f"\n[TRAFIC]")
print(f"  mean                = {mean_req:.2f} req/5min")
print(f"  std                 = {std_req:.2f}")
print(f"  mean + 2*std        = {mean_req + 2*std_req:.2f}")
print(f"  mean + 3*std        = {mean_req + 3*std_req:.2f}")
print(f"  quantile 95%        = {features_df['request_count'].quantile(0.95):.2f}")
print(f"  quantile 97%        = {features_df['request_count'].quantile(0.97):.2f}")
# Au lieu de mean + 2*std (donne ~0%)
# Utiliser un percentile non-zero
non_zero = features_df[features_df["error_rate_5xx"] > 0]["error_rate_5xx"]
SEUIL_ERROR_5XX = non_zero.quantile(0.50)  # médiane des fenêtres avec erreurs
# ou tout simplement fixer un minimum métier
SEUIL_ERROR_5XX = max(mean_err + 2*std_err, 0.05)  # jamais en dessous de 5%
```

```python
# ── ERREURS 5xx ──────────────────────────────────────────
# Distribution bimodale (0 ou 1.0), donc tout seuil > 0 est valide
# On choisit 0.05 (5%) = dès qu'une fenêtre a la moindre erreur serveur
# Justification : 95% des fenêtres = 0 erreur, donc flaguer > 5% est conservateur


# ── TRAFIC ───────────────────────────────────────────────
# mean=1.13, std=0.36 → très peu de variance (dataset de 5000 lignes seulement)
# max observé = 4 req/5min, quantile 97% = 2 → seuil à 3 est raisonnable
# Justification : au-delà de 3 = plus de 2.5× la moyenne, clairement anormal

# More sensitive thresholds → more anomaly examples for the model
SEUIL_ERROR_5XX   = 0.02   # > 2% errors (was 5%)
SEUIL_P95_LATENCE = 1500   # > 1.5s (was 2000ms)
SEUIL_TRAFIC_MULT = 2.0    # 2x average (was 3x)    

# Documentez ainsi dans votre notebook :
print("Justification des seuils :")
print("- Erreurs 5xx : 5% = seuil métier minimal, distribution bimodale (0 ou 100%)")
print("- Latence P95 : 2000ms = mean + 5.6*std, captures only extreme outliers")
print("- Trafic : 3.0 req/5min = ~2.7× la moyenne (1.13), pics clairement anormaux")
```

```python
#Application des labels
cond_error_5xx = features_df["error_rate_5xx"] > SEUIL_ERROR_5XX
cond_latence = features_df["p95_response_time"] > SEUIL_P95_LATENCE
cond_trafic= features_df["request_count"] > (mean_requests * SEUIL_TRAFIC_MULT)

features_df["is_anomaly"] = (cond_error_5xx | cond_latence | cond_trafic).astype(int)

def anomaly_type(row):
    types = []
    if row["error_rate_5xx"] > SEUIL_ERROR_5XX:
        types.append("error_spike")
    if row["p95_response_time"] > SEUIL_P95_LATENCE:
        types.append("latency_spike")
    if row["request_count"] > mean_requests * SEUIL_TRAFIC_MULT:
        types.append("traffic_spike")
    return ",".join(types) if types else "normal"

features_df["anomaly_type"] = features_df.apply(anomaly_type, axis=1)

print("Labels appliqués")
print(f"\nDistribution :")
print(features_df["is_anomaly"].value_counts())
print(f"\nTypes d'anomalies :")
print(features_df["anomaly_type"].value_counts())
```

# **5ter. DIAGNOSTIC DE LA QUALITÉ DU LABEL**

Avant d'adoucir le label, on vérifie d'abord à quel point les anomalies sont
"franches" (loin des seuils) vs "limites" (proches des seuils). Si beaucoup
d'anomalies sont des cas limites, le chevauchement observé dans le modèle
n'est pas un échec du feature engineering mais une conséquence directe
d'un label binaire appliqué à un phénomène continu.

```python
# Distance de chaque ligne aux seuils métier (en % du seuil)
# Une distance proche de 0 = cas limite ; une distance grande = cas franc

dist_error   = (features_df["error_rate_5xx"] - SEUIL_ERROR_5XX) / SEUIL_ERROR_5XX
dist_latence = (features_df["p95_response_time"] - SEUIL_P95_LATENCE) / SEUIL_P95_LATENCE
dist_trafic  = (features_df["request_count"] - mean_requests * SEUIL_TRAFIC_MULT) / (mean_requests * SEUIL_TRAFIC_MULT)

features_df["dist_to_threshold"] = pd.concat([dist_error, dist_latence, dist_trafic], axis=1).max(axis=1)

anomalies_only = features_df[features_df["is_anomaly"] == 1]
print("=== Distribution de la distance au seuil (anomalies uniquement) ===")
print(anomalies_only["dist_to_threshold"].describe())

# Seuil arbitraire pour distinguer "cas limite" (proche de 0) vs "cas franc" (loin)
LIMITE_THRESHOLD = 0.20  # moins de 20% au-dessus du seuil = cas limite
n_limite = (anomalies_only["dist_to_threshold"] < LIMITE_THRESHOLD).sum()
n_total  = len(anomalies_only)
print(f"\nAnomalies 'limites' (< {LIMITE_THRESHOLD*100:.0f}% au-dessus du seuil) : {n_limite}/{n_total} ({n_limite/n_total*100:.1f}%)")
print(f"Anomalies 'franches' (>= {LIMITE_THRESHOLD*100:.0f}% au-dessus du seuil)  : {n_total - n_limite}/{n_total} ({(n_total-n_limite)/n_total*100:.1f}%)")

fig, ax = plt.subplots(figsize=(9, 5))
ax.hist(anomalies_only["dist_to_threshold"], bins=30, color="#e74c3c", alpha=0.7, edgecolor="white")
ax.axvline(LIMITE_THRESHOLD, color="black", linestyle="--", label=f"Seuil cas limite ({LIMITE_THRESHOLD})")
ax.set_xlabel("Distance relative au seuil métier dépassé")
ax.set_ylabel("Nombre d'anomalies")
ax.set_title("Les anomalies sont-elles franches ou limites ?", fontweight="bold")
ax.legend()
plt.tight_layout()
plt.savefig("../report/diagnostic_label_quality.png", bbox_inches="tight")
plt.show()

if n_limite / n_total > 0.3:
    print("\n⚠️  Plus de 30% des anomalies sont des cas limites.")
    print("    → Le chevauchement observé dans le modèle est en grande partie")
    print("      structurel (label binaire sur phénomène continu), pas un échec")
    print("      du feature engineering. Voir section suivante : label fuzzy.")
else:
    print("\n✅  La majorité des anomalies sont franches. Le label est globalement fiable.")
```

## Vérification : corrélation accidentelle ou vraie causalité ?

`is_weekend`, `desktop_ratio`, `get_ratio` sont ressortis en top features
d'importance pour l'Isolation Forest, alors que le label `is_anomaly` ne
dépend que de `error_rate_5xx`, `p95_response_time`, `request_count`.

Hypothèse à tester : avec seulement ~194 anomalies sur 4443 lignes, une
corrélation **accidentelle** (pur bruit d'échantillonnage) entre ces
variables et le label est statistiquement plausible, sans qu'aucune
injection ou biais de génération ne soit en cause.

```python
from scipy import stats

suspect_cols = ["is_weekend", "desktop_ratio", "get_ratio", "auth_ratio", "mobile_ratio"]

print("=== Test de corrélation avec is_anomaly ===\n")
for col in suspect_cols:
    normal_vals  = features_df[features_df["is_anomaly"] == 0][col]
    anomaly_vals = features_df[features_df["is_anomaly"] == 1][col]

    # Test de Mann-Whitney (non paramétrique, adapté aux petits échantillons déséquilibrés)
    stat, pval = stats.mannwhitneyu(normal_vals, anomaly_vals, alternative="two-sided")

    print(f"{col:20s} | moyenne Normal={normal_vals.mean():.3f} | moyenne Anomalie={anomaly_vals.mean():.3f} | p-value={pval:.4f}", 
          "  <-- pas significatif (probablement du bruit)" if pval > 0.05 else "  <-- significatif")

print("\nRappel : p-value > 0.05 signifie qu'on ne peut PAS rejeter l'hypothèse")
print("que la différence observée est due au hasard (taille d'échantillon faible).")
```

```python
# Vérification complémentaire : la corrélation tient-elle sur un sous-échantillon ?
# Si on découpe les anomalies en 2 moitiés aléatoires et que la moyenne de
# is_weekend/desktop_ratio change beaucoup entre les 2 moitiés, c'est le signe
# d'un effet instable = probablement du bruit, pas un vrai signal.

rng = np.random.RandomState(0)
anomalies_idx = features_df[features_df["is_anomaly"] == 1].index.to_numpy()
anomalies_idx = features_df[features_df["is_anomaly"] == 1].index.to_numpy().copy()
rng.shuffle(anomalies_idx)
half1, half2 = anomalies_idx[:len(anomalies_idx)//2], anomalies_idx[len(anomalies_idx)//2:]

print("=== Stabilité de la moyenne entre 2 sous-échantillons aléatoires d'anomalies ===\n")
for col in suspect_cols:
    m1 = features_df.loc[half1, col].mean()
    m2 = features_df.loc[half2, col].mean()
    diff_pct = abs(m1 - m2) / (abs(m1) + 1e-6) * 100
    print(f"{col:20s} | moitié 1={m1:.3f} | moitié 2={m2:.3f} | écart relatif={diff_pct:.0f}%")

print("\nUn écart relatif élevé (> 30-40%) entre les 2 moitiés indique une")
print("estimation instable = peu fiable avec seulement ~194 anomalies au total.")
```

## Adoucir le label : catégorie intermédiaire "Warning"

Au lieu d'un label strict Normal/Anomalie, on introduit une 3e catégorie
pour les cas proches du seuil. Ces cas seront exclus du calcul des métriques
(F1, seuil optimal) pour ne pas brouiller l'évaluation avec des exemples
ambigus, tout en restant visibles dans les données pour analyse manuelle.

```python
def label_with_warning_zone(row, marge=0.20):
    """
    3 catégories :
      - 0 = Normal       : aucun critère dépassé
      - 1 = Anomalie      : au moins un critère dépassé largement (> marge au-dessus du seuil)
      - 2 = Warning       : un critère est dépassé mais de peu (cas limite, < marge)
    """
    error_ratio   = row["error_rate_5xx"] / SEUIL_ERROR_5XX if SEUIL_ERROR_5XX > 0 else 0
    latence_ratio = row["p95_response_time"] / SEUIL_P95_LATENCE
    trafic_ratio  = row["request_count"] / (mean_requests * SEUIL_TRAFIC_MULT)

    max_ratio = max(error_ratio, latence_ratio, trafic_ratio)

    if max_ratio < 1.0:
        return 0  # Normal : sous tous les seuils
    elif max_ratio < 1.0 + marge:
        return 2  # Warning : juste au-dessus, cas limite
    else:
        return 1  # Anomalie : nettement au-dessus

features_df["label_3class"] = features_df.apply(label_with_warning_zone, axis=1)

print("=== Distribution du label à 3 classes ===")
print(features_df["label_3class"].map({0: "Normal", 1: "Anomalie", 2: "Warning"}).value_counts())

# Pour l'évaluation du modèle : label binaire "propre", Warning exclu
features_df["is_anomaly_clean"] = features_df["label_3class"].map({0: 0, 1: 1, 2: np.nan})

n_warning = (features_df["label_3class"] == 2).sum()
print(f"\n{n_warning} fenêtres en zone Warning seront exclues du calcul du F1/seuil")
print("(mais restent dans le dataset complet pour l'inspection manuelle)")
```

# **5bis. FEATURE ENGINEERING NON-SUPERVISÉ**

Les features brutes (`avg_response_time`, `error_rate_5xx`, etc.) sont de bons signaux,
mais pris isolément, l'Isolation Forest a du mal à isoler les anomalies dans un espace
à 33 dimensions bruité.

**Important** : ces nouvelles features sont construites UNIQUEMENT à partir des variables
d'entrée (latence, erreurs, trafic, etc.) — **jamais** à partir de `is_anomaly`/`anomaly_type`,
qui restent réservés à l'évaluation finale du modèle.

Objectif : combiner les signaux existants pour créer des axes où les comportements
anormaux sont plus facilement isolables géométriquement (ce qui est le principe même
de l'Isolation Forest : un point anormal nécessite moins de splits pour être isolé).

```python
# === FEATURES COMPOSITES (combinaisons de signaux bruts) ===
# Construites uniquement à partir des features déjà calculées, sans utiliser is_anomaly

# 1. Score de stress serveur : latence amplifiée par le taux d'erreur
#    Un serveur lent ET en erreur est pire que la somme des deux pris séparément
features_df["stress_score"] = features_df["avg_response_time"] * (1 + features_df["error_rate_5xx"])

# 2. Dégradation de la qualité de service : latence relative au taux de succès
features_df["quality_degradation"] = features_df["avg_response_time"] / (features_df["success_rate"] + 0.01)

# 3. Spike de latence : écart entre la queue de distribution (P99) et la médiane
#    Un écart important indique un sous-ensemble de requêtes très lentes (signe d'anomalie)
features_df["latency_spike"] = features_df["p99_response_time"] - features_df["median_response_time"]

# 4. Sévérité des erreurs : proportion d'erreurs serveur (5xx) parmi toutes les erreurs
total_errors = features_df["error_rate_5xx"] + features_df["error_rate_4xx"] + 1e-6
features_df["error_severity_ratio"] = features_df["error_rate_5xx"] / total_errors

# 5. Coefficient de variation de la latence : volatilité relative
#    Une forte volatilité (CV élevé) peut signaler de l'instabilité serveur
features_df["response_cv"] = features_df["std_response_time"] / (features_df["avg_response_time"] + 1e-6)

# 6. Pression de charge : volume de requêtes pondéré par la latence
features_df["load_pressure"] = features_df["request_count"] * features_df["avg_response_time"]

# 7. Transformations log pour réduire l'effet des outliers extrêmes
#    L'Isolation Forest est sensible aux échelles très étirées par quelques valeurs extrêmes
features_df["log_avg_response"] = np.log1p(features_df["avg_response_time"])
features_df["log_p99_response"] = np.log1p(features_df["p99_response_time"])
features_df["log_std_response"] = np.log1p(features_df["std_response_time"])
features_df["log_load_pressure"] = np.log1p(features_df["load_pressure"])

# 8. Écart par rapport au comportement récent du même serveur (déjà partiellement présent
#    via response_vs_rolling) — on ajoute la version normalisée et le z-score local
features_df["response_vs_rolling_abs"] = features_df["response_vs_rolling"].abs()

# Z-score de la latence par serveur (utile car chaque serveur a son propre régime normal)
features_df["response_zscore_per_server"] = (
    features_df.groupby("server_id")["avg_response_time"]
    .transform(lambda x: (x - x.mean()) / (x.std() + 1e-6))
)

print("Nouvelles features composites créées :")
new_cols = [
    "stress_score", "quality_degradation", "latency_spike", "error_severity_ratio",
    "response_cv", "load_pressure", "log_avg_response", "log_p99_response",
    "log_std_response", "log_load_pressure", "response_vs_rolling_abs",
    "response_zscore_per_server"
]
for c in new_cols:
    print(f"  - {c}")
print(f"\nShape features_df : {features_df.shape}")
```

### Features de sévérité continue (point 3)

Au lieu de laisser l'Isolation Forest redécouvrir seul où se trouvent les
seuils métier dans un espace bruité, on lui fournit directement une mesure
de **distance signée et continue** par rapport à ces seuils. Une valeur
négative = sous le seuil (normal), positive = au-dessus (anormal), et
l'amplitude indique la sévérité. Ce signal monotone est beaucoup plus
facile à exploiter géométriquement qu'une variable brute non recentrée.

```python
# Distance normalisée par rapport aux 3 seuils métier (capture la sévérité continue)
# Valeur < 0  -> sous le seuil (normal)
# Valeur > 0  -> au-dessus du seuil (anormal), amplitude = sévérité

features_df["severity_latency"] = (
    (features_df["p95_response_time"] - SEUIL_P95_LATENCE) / SEUIL_P95_LATENCE
)
features_df["severity_error"] = (
    (features_df["error_rate_5xx"] - SEUIL_ERROR_5XX) / max(SEUIL_ERROR_5XX, 1e-6)
)
features_df["severity_traffic"] = (
    (features_df["request_count"] - mean_requests * SEUIL_TRAFIC_MULT)
    / (mean_requests * SEUIL_TRAFIC_MULT)
)

# Score de sévérité globale : le pire des 3 critères (cohérent avec la logique OR du label)
features_df["max_severity"] = features_df[
    ["severity_latency", "severity_error", "severity_traffic"]
].max(axis=1)

# Score de sévérité combinée : somme des sévérités positives uniquement
# (capture les cas où PLUSIEURS critères sont dégradés simultanément, même légèrement,
#  ce qui est souvent plus révélateur qu'un seul critère très dégradé)
severity_cols = ["severity_latency", "severity_error", "severity_traffic"]
features_df["combined_severity"] = features_df[severity_cols].clip(lower=0).sum(axis=1)

print("Features de sévérité créées : severity_latency, severity_error, severity_traffic, max_severity, combined_severity")
print(features_df[["max_severity", "combined_severity"]].describe())
```

### Vérification visuelle (à titre indicatif uniquement)

On regarde si ces nouvelles features séparent mieux les classes — **uniquement pour

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
check_cols = ["stress_score", "quality_degradation", "latency_spike",
              "error_severity_ratio", "response_cv", "response_zscore_per_server"]

for ax, col in zip(axes.flatten(), check_cols):
    normal = features_df[features_df["is_anomaly"] == 0][col]
    anomaly = features_df[features_df["is_anomaly"] == 1][col]
    ax.hist(normal, bins=40, alpha=0.6, color="#3498db", label="Normal", density=True)
    ax.hist(anomaly, bins=20, alpha=0.7, color="#e74c3c", label="Anomalie", density=True)
    ax.set_title(col, fontsize=10, fontweight="bold")
    ax.legend(fontsize=8)

plt.suptitle("Séparation des features composites (contrôle qualité)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../report/feature_engineering_check.png", bbox_inches="tight")
plt.show()
print("Figure sauvegardée : feature_engineering_check.png")
```

# **6. Normalisation StandardScaler**

```python
from sklearn.preprocessing import StandardScaler
import pickle

# Exclure les colonnes non-numériques, les métadonnées, et les colonnes de label/diagnostic
# Toutes les colonnes liées au label (is_anomaly, label_3class, is_anomaly_clean,
# dist_to_threshold) sont gardées à part : UNIQUEMENT pour l'évaluation finale,
# jamais utilisées comme input du modèle (apprentissage non-supervisé)
EXCLUDE_COLS = [
    "server_id", "timestamp",
    "is_anomaly", "anomaly_type", "label_3class", "is_anomaly_clean", "dist_to_threshold"
]

FEATURES_ML = [
    col for col in features_df.columns
    if col not in EXCLUDE_COLS
    and features_df[col].dtype in ["float64", "int64", "float32", "int32"]
]

print(f"Features détectées automatiquement : {len(FEATURES_ML)}")
print(FEATURES_ML)

X = features_df[FEATURES_ML].fillna(0)
y = features_df["is_anomaly"]                  # label binaire original, pour évaluation
y_clean = features_df["is_anomaly_clean"]       # label sans la zone Warning, pour évaluation affinée

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=FEATURES_ML)

print(f"Normalisation appliquée")
print(f"Shape X_scaled : {X_scaled.shape}")
print(f"Shape y : {y.shape}")
```

```python
X_scaled.to_csv("../data/X_scaled.csv", index=False)
y.to_csv("../data/y_labels.csv", index=False)
y_clean.to_csv("../data/y_labels_clean.csv", index=False)   # label sans la zone Warning
features_df.to_csv("../data/features_ml.csv", index=False)  # contient is_anomaly, label_3class, severity_*, dist_to_threshold

with open("../data/scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("Fichiers sauvegardés :")
print("  - X_scaled.csv         => features normalisées pour le ML")
print("  - y_labels.csv         => labels anomalies (binaire, incluant les cas Warning)")
print("  - y_labels_clean.csv   => labels anomalies, NaN sur la zone Warning (évaluation affinée)")
print("  - features_ml.csv      => données complètes avec contexte, severity_*, label_3class")
print("  - scaler.pkl           => StandardScaler pour l'inférence")
```
